## Patterns for Min sum, cost , path patterns

*Statement* Given a target find minimum (maximum) cost / path / sum to reach the target. 

*Problem Link*: https://leetcode.com/problems/triangle/submissions/?envType=list&envId=55ac4kuc



```
class Solution:
    def minCostClimbingStairs(self, cost: List[int]) -> int:
        res = 0
        for i  in range(len(cost)-1):
            res += min(cost[i] , cost[i+1])
        
        return res
```
